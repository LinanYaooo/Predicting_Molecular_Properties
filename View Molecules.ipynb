```python
import ase
import pandas as pd
import random

from ase import Atoms
import ase.visualize
```

[Knowledges](https://www.khanacademy.org/science/organic-chemistry/spectroscopy-jay#proton-nmr)

```python
path = '../dataset/structures.csv'
data = pd.read_csv(path)



def view(molecule):
    # Select a molecule
    mol = data[data['molecule_name'] == molecule]
    
    # Get atomic coordinates
    xcart = mol.iloc[:, 3:].values
    print(xcart.shape)
    print('structure:')
    print(xcart)
    # Get atomic symbols

    symbols = mol.iloc[:, 2].values
    print('elements:')
    print(symbols)
    
    # Display molecule
    system = Atoms(positions=xcart, symbols=symbols)
    print('Molecule Name: %s.' %molecule)
    return ase.visualize.view(system, viewer="x3d")
```

```python
random_molecule = random.choice(data['molecule_name'].unique())
view(random_molecule)
```
